VIRGINIA COVID-19 STATS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
death_dates = ['3/14', '3/15', '3/16', '3/17', '3/18', '3/19', '3/20', '3/21', '3/22', '3/23', '3/24', '3/25', '3/26', '3/27', '3/28', '3/29', '3/30', '3/31', 
               '4/1', '4/2', '4/3', '4/4', '4/5', '4/6', '4/7', '4/8', '4/9', '4/10',
               '4/11', '4/12', '4/13', '4/14', '4/15', '4/16', '4/17', '4/18', '4/19', '4/20',
               '4/21', '4/22', '4/23', '4/24', '4/25', '4/26', '4/27', '4/28', '4/29', '4/30',]
```

```python
deaths_per_day = [1, 0, 1, 0, 0, 0, 0, 0, 1, 4, 0, 2, 0, 5, 3, 5, 3, 2, 7, 7, 5, 6]
```

```python
Virginia_deaths = pd.DataFrame(list(zip(death_dates, deaths_per_day)))
```

```python
Virginia_deaths.columns = ['death_dates', 'deaths']
```

```python
Virginia_deaths.set_index('death_dates')
```

```python
plt.figure(figsize=(12, 9))
death_daily = sns.barplot(x='death_dates', y='deaths', data=Virginia_deaths)
for p in death_daily.patches:
    death_daily.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0,10), textcoords='offset points')
plt.xticks(rotation=60)
plt.title('Virginia COVID-19 deaths by day')
plt.show()
```

```python
plt.figure(figsize=(12, 9))
death_plot = sns.barplot(x='death_dates', y=Virginia_deaths.deaths.cumsum(), data=Virginia_deaths)
for p in death_plot.patches:
    death_plot.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0,10), textcoords='offset points')
plt.xticks(rotation=60)
plt.title('Virginia cumulative COVID-19 deaths')
#plt.savefig('Viginia cumulative COVID-19 deaths as of March 29')
plt.show()
```

```python
new_reports_per_day = [1, 1, 3, 3, 1, 8, 13, 11, 4, 6, 15, 10, 16, 22, 39, 69, 32, 36, 101, 69, 144, 135, 151, 130, 230, 234, 222, 306, 395]
```

```python
Virginia_cases = pd.DataFrame(list(zip(new_reports_dates, new_reports_per_day)))
```

```python
Virginia_cases.columns = ['new_reports_dates', 'new_reports_per_day']
```

```python
Virginia_cases['double_rate_cases'] = 70 / (Virginia_cases['new_reports_per_day'].cumsum().pct_change() * 100)
```

```python
Virginia_cases.set_index('new_reports_dates')
```

```python
plt.figure(figsize=(12, 9))
reports_daily = sns.barplot(x='new_reports_dates', y='new_reports_per_day', data=Virginia_cases)
for p in reports_daily.patches:
    reports_daily.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0,10), textcoords='offset points')
plt.xticks(rotation=60)
plt.title('Virginia COVID-19 reports by day')
plt.show()
```

```python
plt.figure(figsize=(12, 9))
cases_plot = sns.barplot(x='new_reports_dates', y=Virginia_cases.new_reports_per_day.cumsum(), data=Virginia_cases)
for p in cases_plot.patches:
    cases_plot.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0,10), textcoords='offset points')
plt.xticks(rotation=60)
plt.title('Virginia cumulative COVID-19 reported cases')
#plt.savefig('Viginia cumulative COVID-19 reports as of March 30')
plt.show()
```

```python
plt.figure(figsize=(12, 9))
VA_cases_double_rate = sns.barplot(x='new_reports_dates', y='double_rate_cases', data=Virginia_cases)
for p in VA_cases_double_rate.patches:
    VA_cases_double_rate.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0,10), textcoords='offset points')
plt.xticks(rotation=60)
plt.title('Virginia COVID-19 doubling rate')
plt.show()
```
